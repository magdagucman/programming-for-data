# Introduction

The investigation of existing international data relating to suicides juxtaposed with other variables that might be linked to these suicides can potentially provide useful insights about suicide causality seen as a national issue and, therefore, become a scientific basis for national suicide prevention programmes. 

Using records taken from UN online databases, the study will investigate how well the number of suicides can be predicted by various other factors: unemployment, poverty, inequality, accessibility of psychiatrists, and accessibility of psychiatric hospitals. The study will also investigate whether higher gender inequality can be connected with higher percentage of suicides among women.

The study will report on various correlations, as well as present a multidimensional linear regression model (see 'Hypotheses' section for details), to look into potentially complex causality of suicides and to raise the topic of suicide prevention needing to be focused on satisfying basic human needs as well as providing access to (ideally free) mental healthcare.


## Links to data:

The data is available for years 1991, 1995, 2000, 2005, 2010-2019 for 180 countries associated with their HDI rank. Link: https://raw.githubusercontent.com/magdagucman/programming-for-data/main/Data%20Sets/Unemployment.csv

* Population living below income poverty line, national poverty line (%): contains records of percentage of the population living below the national poverty line, which is the poverty line deemed appropriate for a country by its authorities. The data is available for a range of years between 2008 and 2019 (most recent data available during the specified period) for 130 countries associated with HDI rank. Link: https://raw.githubusercontent.com/magdagucman/programming-for-data/main/Data%20Sets/Poverty.csv

* Coefficient of human inequality: contains records of the coefficient calculated as the arithmetic mean of the values in inequality in life expectancy, inequality in education and inequality in income for years 2010-2019 for 167 countries (some values missing in various years). Countries are associated with their HDI rank. Link: https://raw.githubusercontent.com/magdagucman/programming-for-data/main/Data%20Sets/Inequality.csv

* Psychiatrists working in mental health sector (per 100,000): contains records of psychiatrists per 100000 people for year 2011 for 180 countries. Link: https://raw.githubusercontent.com/magdagucman/programming-for-data/main/Data%20Sets/Psychiatrists.csv

* Mental hospitals (per 100,000): contains records of mental hospitals per 100000 people for year 2011 for 173 countries. Link: https://raw.githubusercontent.com/magdagucman/programming-for-data/main/Data%20Sets/Hospitals.csv

* Gender Inequality Index: contains records for years 1995, 2000, 2005, 2010-2019 for 162 countries and associated HDI rank.  https://raw.githubusercontent.com/magdagucman/programming-for-data/main/Data%20Sets/Gender%20Inequality%20Index%20(GII).csv

## Data Cleaning & Wrangling

In the below code I will address the issues identified above.

I will start with identifying and removing all-empty columns.

```python
# create a function to remove all-empty columns and rows from the listed dataframes
def all_empty_remover(df_list):
    num = 0
    
    # iterate through the dataframes in the list
    for dataframe in df_list:
        
        # drop the all-empty columns
        dataframe = dataframe.dropna(how='all', axis=1)
        
        # overwrite the dataframe
        df_list[num] = dataframe
        num+=1
    return df_list

# run the all_empty_remover function 
df_list = all_empty_remover(df_list)

# display the dataframes again, to check visually whether the the above function worked
displayer(df_list)
```

### Outcomes summary

All entirely empty columns are now removed. 

---

I will now decide on the columns I want to work with. 

I will focus on the year 2010, as it's available for the suicides, and at the same time data about psychiatrists and mental hospitals is available only for year 2011, so it is the closest I can get. One more problematic dataset is the one referring to poverty, as it does not state which year the data is from, so any results concerning poverty data will be approached with caution.

I will now remove the unnecessary columns from the dataframes, to leave only the data I want to use in my analyses - country and values from the appropriate year (2010 or 2011).

```python
# define a function to iterate through columns and drop the ones that won't be used
def choose_columns(df_list):
    num = 0
    
    # iterate through dataframes
    for df in df_list:
        column_list=[]
        
        # iterate through columns
        for column in df.columns:
            
            # check condition - if column is not on the list, append it to a list of columns to be dropped
            if column not in ['Country', 'Country or Area', '2010', '2008-2019', 'Value']:
                column_list.append(column)
                new = df.drop(columns=column_list)
                
                # overwrite the dataframe
                df_list[num] = new
        num+=1
    return df_list

df_list = choose_columns(df_list)

# display the dataframes to check whether the function above worked correctly, leaving only appropriate columns
displayer(df_list)
```

### Outcomes Summary

The unnecessary columns have now been removed, leaving only the columns containing the name of the country together with values from appropriate years - 2010/11.

---

I observed that some rows that do not contain any values, contain '..' instead, and so are not recognised as empty rows. Below I will make sure to remove those rows.

```python
# create a function to remove the rows that contain '..' instead of actual values
def dotdot_remover(df_list):
    
    # iterate through dataframes
    for df in df_list:
        
        # iterate through each row
        for row in df:
            
            # create a condition - if a '..' in not empty (exists)
            if df[df[row] == '..'].empty == False:
                
                # create a list of row idices that contain the '..' value 
                index_names = df[df[row] == '..'].index
                
                # drop the rows on the list above, overwrite
                df = df.drop(index_names, inplace = True)
    return df_list

df_list = dotdot_remover(df_list)

# rename the appropriate columns to avoid any confusion after merging
df_list[0] = df_list[0].rename(columns={"2010": "suicide_female"})
df_list[1] = df_list[1].rename(columns={"2010": "suicide_male"})
df_list[2] = df_list[2].rename(columns={"2010": "unemployment"})
df_list[3] = df_list[3].rename(columns={"2008-2019": "poverty"})
df_list[4] = df_list[4].rename(columns={"2010": "inequality"})
df_list[5] = df_list[5].rename(columns={"Country or Area": "Country", "Value": "psychiatrists"})
df_list[6] = df_list[6].rename(columns={"Country or Area": "Country", "Value": "mental_hospitals"})
df_list[7] = df_list[7].rename(columns={"2010": "gender_inequality"})

# display to check if above executed correctly 
displayer(df_list)
```

Displaying dataframes as a way of testing showed that '..' value rows were removed and names of columns were changed successfully.

---

The next step is to merge the dataframes, to have the data of interest easily accessible and neatly organised within one dataframe. To do that, I will write a merging function.

```python
# As I initially run my merger function, I encountered an error which caused the final table to appear empty
# Quick test of country names showed that lengths were not the same, meaning there were whitespaces needing to be deleted
for df in df_list:
    df['Country'] = df['Country'].str.strip()

def merger(df_list, col):
    merged = df_list[0]
    for i in range(len(df_list)-1):
        merged = merged.merge(df_list[i+1], left_on=col, right_on=col)
    return merged
    

merged = merger(df_list, 'Country')

# Before displaying and analysing the summary statistics, I will ensure that all the values are of correct type. 
# This means that for the 'inequality' and 'gender_inequality' columns I need to change type from object to float64.

merged['inequality'] = merged['inequality'].astype('float64')
merged['gender_inequality'] = merged['gender_inequality'].astype('float64')

display(merged.info())
```

Lastly, I will create 2 new columns - suicide_total and suicide_ratio, to later be used in the analyses.

```python
merged['suicide_total'] = merged['suicide_female'] + merged['suicide_male']
merged['suicide_ratio'] = merged['suicide_female'] / merged['suicide_male']

display(merged.info())
display(merged.head(10))
```

### Outcomes Summary

Display showed that the merging function worked successfully and all the data is now stored within one dataframe. Two new columns have successfully been added to the dataframe as well.

## Assumptions

I will firstly use boxplots to identify outliers. To do that, I will write a function outlier_identifier.

```python
import seaborn as sns
import matplotlib.pyplot as plt

def outlier_identifier(df):
    for column in df:
        try:
            sns.boxplot(y=df[column], data=merged)
            plt.show()
        except:
            pass

outlier_identifier(merged)
```

The boxplots show that there are many outliers in various columns. I will ignore suicide females and males, and focus on the total amount of suicides, as I won't be analysing these separately.

* Suicide total shows 4 outliers
* Suicide ratio shows 8 outliers
* mental_hospitals shows 6 outliers
* psychiatrists shows 7 outliers
* unemployment shows 3 outliers

The next step is to write an outlier removing function and remove the outliers.

```python
def remove_outliers(df, column):
    q3 = df[column].quantile(0.75)
    q1 = df[column].quantile(0.25)
    iqr = q3 - q1
    upper_limit = (q3 + 1.5*iqr)
    lower_limit = (q1 - 1.5*iqr)
    
    # Use loc to access rows by boolean array
    # Save a group of rows into df_normal (conditional is true for values bigger than lower limit and lower than upper limit)
    df_normal = df.loc[(df[column].values > lower_limit) & (df[column].values < upper_limit)]
    
    return df_normal

# remove the outliers column by column
merged_noout = remove_outliers(merged, 'unemployment')

merged_noout = remove_outliers(merged_noout, 'psychiatrists')

merged_noout = remove_outliers(merged_noout, 'mental_hospitals')

merged_noout = remove_outliers(merged_noout, 'suicide_total')

merged_noout = remove_outliers(merged_noout, 'suicide_ratio')

# display the info to test whether the outliers have been removed
merged_noout.info()
```

### Outcomes Summary

The outliers have now been removed, leaving data from 52 countries in the dataframe.

I will now display the mean, median and histograms for all columns in the dataframe, to test the assumption of normality. To do this, I will write a function which will iterate through the columns and display the aforementioned data.

```python
def is_normal(df):
    
    for column in df:
        try:
            display(df[column].mean())
            display(df[column].median())
            sns.displot(df[column])
            plt.show()
        except:
            pass

is_normal(merged_noout)
```

The mean and median for inequality are similar and the histogram resembles bell curve.

The mean and median for psychiatrists are not similar and the histogram doesn't resemble bell curve, suggesting that data distribution is not normal in this case.

The mean and median for mental hospitals are similar but the histogram does not resemble bell curve, duggesting the not normal data distribution.

The mean and median for gender inequality are similar and the historgam resembles bell curve suggesting normal distribution.

The mean and median for total suicides are similar and histogram might resemble bell curve - possibly the distribution is close enough to normal.

## Linear regression

In sections below I will test hypotheses 1-6 using linear regression.

### H1

There is a positive, statistically significant correlation between the total number of suicides and unemployment, measured as a total percentage of labour force.

```python
from scipy.stats import linregress

def regression(df, x, y):
    x = df[x]
    y = df[y]

    regression = linregress(x,y)

    display(regression, regression.slope, regression.intercept, regression.rvalue, regression.pvalue)

    sns.regplot(y=y, x=x, data=df)

regression(merged_noout, 'unemployment', 'suicide_total')
```

The results show that there is no significant relation between unemployment and total suicides. The r value is close to 0, p close to 1.

### H2

There is a positive, statistically significant correlation between the total number of suicides and poverty, measured by percentage of population living below national income poverty line.

```python
regression(merged_noout, 'poverty', 'suicide_total')
```

The results show that poverty can be a predictor of the total number of suicides (p < 0.05, so the model is significant), accounting for 28% of variance. The higher the poverty in the country, the higher the total amount of suicides.

### H3

There is a positive, statistically significant correlation between the total number of suicides and inequality, measured by the Coefficient of Human Inequality.

```python
regression(merged_noout, 'inequality', 'suicide_total')
```

The results show no significant correlation between the variables (p > 0.05).

### H4

There is a negative, statistically significant correlation between the total number of suicides and accessibility of psychiatrists, measured by the number of psychiatrists working in mental health sector per 100000 people.

```python
regression(merged_noout, 'psychiatrists', 'suicide_total')
```

The results show no significant correlation between the variables (p > 0.05).

### H5

There is a negative, statistically significant correlation between the total number of suicides and accessibility of mental hospitals, measured by the number of mental hospitals per 100000 people.

```python
regression(merged_noout, 'mental_hospitals', 'suicide_total')
```

The results show no significant correlation between the variables (p > 0.05).

### H6

There is a positive, statistically significant relation between the ratio of female to male suicides and gender inequality index.

```python
regression(merged_noout, 'gender_inequality', 'suicide_ratio')
```
